```python
df.groupby('Category')["Sales"].sum().plot(kind='bar')

plt.title("sales by category")
plt.xlabel("category")
plt.ylabel("sales")
plt.show()
```

```python
df.groupby("Region")["Profit"].sum().plot( kind ='bar')

plt.title("profit by region")
plt.xlabel("region")
plt.ylabel("profit")
plt.show()
```

```python
df.groupby('year')['Sales'].sum().plot(kind='line')

plt.title("Monthly Sales")
plt.show()
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().head(10)

top_products.plot(kind="line")
plt.title("top 10 products")
plt.show()
```
